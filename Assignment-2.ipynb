```python
import pandas as pd
df=pd.read_csv('supershops.csv')
```

```python
df.head()
```

# For Standardization

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
```

```python
scaled=ss.fit(df[['Marketing Spend']])
```

```python
df['Marketing Spend']=ss.transform(df[['Marketing Spend']])
```

```python
df.head()
```

```python
df['Administration']=ss.transform(df[['Administration']])
```

```python
df.head()
```

```python
df['Transport']=ss.transform(df[['Transport']])
```

```python
df.head()
```
